```python
df6 = pandas.read_csv("http://pythonhow.com/supermarkets.csv")  # to open online file
df6
```

```python
df7 = df6.set_index("Address")   # save table with index "address" to a new variable
df7
```

```python
# label based indexing:
df7.loc["735 Dolores St" : "3995 23rd St", "City" : "Country"]   # slice of table with range of rows and range of columns
```

```python
# getting a single cell:
df7.loc["332 Hill St", "Name"]
```

```python
# getting 1 row to a list:
list(df7.loc["551 Alvarado St", : ])  # excludes the index cell!
```

```python
# position based indexing:
df7.iloc[1:4, 1:4]   # range here is upperbound exclusive!
```

```python
# getting a single cell:
df7.iloc[2, 4]   # range excludes the index cell!
```

```python
# getting 1 row to a list:
list(df7.iloc[5, : ])  # excludes the index cell!
```

```python
# combined indexing (Deprecated):
df7.ix[2, "Name"]   # can combine position and label, also can use range similarily as .loc and .iloc
```

```python
df7
```

```python
# label based deleting - rows
df7.drop("3995 23rd St", 0)   # 0 means a row
```

```python
# label based deleting - colums
df7.drop("City", 1)   # 1 means a column
```

```python
# index based deleting - rows
df7.drop(df7.index[0:3], 0)
```

```python
# index based deleting - columns
df7.drop(df7.columns[0:3], 1)
```

```python
df7
```

```python
# adding a new column: "Continent" and add same value to all rows
# this is an IMPLACE method, so it will update and save the dataframe
df7["Continent"] = df7.shape[0] * ["North America"]  # 0th element of shape tuple is the number of rows, so we multiply
```

```python
df7
```

```python
# adding a new row
# first need to transpose the dataframe (change rows to columns and v.v.) to a new variable
# then actually add a new column with the previous method
# and then transpose it back to original form
df7_transp = df7.T
df7_transp
```

```python
df7_transp["My Address"] = [7, "My City", "My State", "My Country", "My Shopname", 17, "My Continent"]
df7 = df7_transp.T
df7
```
